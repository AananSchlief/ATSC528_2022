```python
### Import Required Modules (shouldn't need to change) ###
import numpy as np                 #numpy for math
import matplotlib.pyplot as plt    #matplotlib for plotting
import cartopy.crs as ccrs         #cartopy for plotting on map
import cartopy.feature as cfeature #cartopy basic shapefiles
```

```python
#Get constants from pdf 
a = 0.25
b = 4.0
f = 8.0
g = 1.0
dt = 4/(24*5)  
tstep = 120/5/dt
tstep = int(tstep)

## empty arrays
x1 = np.empty(tstep)
y1 = np.empty(tstep)
z1 = np.empty(tstep)

x2 = np.empty(tstep)
y2 = np.empty(tstep)
z2 = np.empty(tstep)

## Set initial condtions for the arrays
x1i = 2.5
y1i = 1.0
z1i = 0.0

x2i = 2.49
y2i = 1.0
z2i = 0.0 

## Put initial conditions into arrays
x1[0] = x1i
y1[0] = y1i
z1[0] = z1i

x2[0] = x2i
y2[0] = y2i
z2[0] = z2i
```

```python
#set up for loop to calculate x1,y1,z1,x2,y2,z2
for i in range(1, tstep):
    #do first set of initial conditons 
    x1_calc = x1[i-1] + dt*(-y1[i-1]**2 - z1[i-1]**2 - (a*x1[i-1]) + (a*f))
    y1_calc = y1[i-1] + dt*(x1[i-1] * y1[i-1] - (b*x1[i-1]*z1[i-1]) - y1[i-1] + g)
    z1_calc = z1[i-1] + dt*((b*x1[i-1]*y1[i-1]) + (x1[i-1]*z1[i-1]) - z1[i-1])
    
    # do second set of initial conditions 
    x2_calc = x2[i-1] + dt*(-y2[i-1]**2 - z2[i-1]**2 - (a*x2[i-1]) + (a*f))
    y2_calc = y2[i-1] + dt*(x2[i-1] * y2[i-1] - (b*x2[i-1]*z2[i-1]) - y2[i-1] + g)
    z2_calc = z2[i-1] + dt*((b*x2[i-1]*y2[i-1]) + (x2[i-1]*z2[i-1]) - z2[i-1])
    
    x1[i] = x1_calc
    y1[i] = y1_calc
    z1[i] = z1_calc
    
    x2[i] = x2_calc
    y2[i] = y2_calc
    z2[i] = z2_calc
```

```python
## plot X values 
plt.plot(x1, color = 'black', label = 'x1')
plt.plot(x2, color = 'red', label = 'x2')
plt.legend()
plt.title('X Chaos Simulation')
plt.savefig('xchaos.png')
plt.show
```

```python
## plot y values
plt.plot(y1, color = 'black', label = 'y1')
plt.plot(y2, color = 'red', label = 'y2')
plt.legend()
plt.title('Y Chaos Simulation')
plt.savefig('ychaos.png')
plt.show
```

```python
## plot z values 
plt.plot(z1, color = 'black', label = 'z1')
plt.plot(z2, color = 'red', label = 'z2')
plt.legend()
plt.title('Z Chaos Simulation')
plt.savefig('zchaos.png')
plt.show
```
